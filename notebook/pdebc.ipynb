```python
import pylab as pl
from numpy import *
from scipy.misc import *

def explicitBezier(p, t):
    np = len(p)
    n = np - 1
    
    sx = 0
    sy = 0
    for i in range(np):
        bi = comb(n,i)
        t1 = (1-t)**(n-i)
        t2 = t**i
        t3 = bi*t1*t2
        sx = sx + t3*p[i][0]
        sy = sy + t3*p[i][1]
    return (sx,sy)
```

```python
matplotlib.rcParams['figure.figsize'] = (8,5)
rcParams['savefig.dpi'] = 150

def plotBC(ps=None, tn=0, dp=None, pop=None):
    pl.clf()
    min_x = 0
    min_y = 0
    max_x = 1
    max_y = 1
    dd = 5
    if ps != None:
        x = []
        y = []
        for t in range(0,tn):
            (bx,by) = explicitBezier(ps, float(t)/float(tn-1))
            x = x + [bx]
            y = y + [by]

        px = zip(*ps)[0]
        py = zip(*ps)[1]
    
        pl.plot(x,y, label='Bezier Curve')
        pl.plot(px,py), '--')
        pl.plot(px,py, 'rs', label='Bezier Curve Control Points')
    
        min_x = min(min(x),min(px))-dd
        min_y = min(min(y),min(py))-dd
        max_x = max(max(x),max(px))+dd
        max_y = max(max(y),max(py))+dd

    if pop != None:
        pl.plot(zip(*pop[0])[0],zip(*pop[0])[1], 'x', label="P1's Population", color='y')
        pl.plot(zip(*pop[1])[0],zip(*pop[1])[1], "+", label="P2's Population", color='#333333')
        if ps == None and dp == None:
            min_x = min(zip(*pop[0])[0])-dd
            min_y = min(zip(*pop[0])[1])-dd
            max_x = max(zip(*pop[0])[0])+dd
            max_y = max(zip(*pop[0])[1])+dd
    if dp != None:
        pl.plot(zip(*dp)[0],zip(*dp)[1], '.', label="Data Points", color='k')
        min_x = min(min(zip(*dp)[0])-dd,min_x)
        min_y = min(min(zip(*dp)[1])-dd,min_y)
        max_x = max(max(zip(*dp)[0])+dd,max_x)
        max_y = max(max(zip(*dp)[1])+dd,max_y)
    
    
    pl.axis([min_x, max_x, min_y, max_y])
    pl.legend()
```

```python
p1 = (5,5)
p2 = (5,20)
p3 = (15,-20)
p4 = (25, 0)

p = [p1,p2,p3,p4]

tn = 20
x = []
y = []
for t in range(0,tn):
    (bx,by) = explicitBezier(p, float(t)/float(tn-1))
    x = x + [bx]
    y = y + [by]

dpx = x + []
dpy = y + []
dp = zip(dpx,dpy)

#plotBC(None, 20, dp)
#plotBC(p,100,dp)
```
